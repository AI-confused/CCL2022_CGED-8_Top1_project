```python
# 处理lan8数据
import os
ROOT = os.path.abspath(os.path.dirname(os.getcwd()))
print(ROOT)
try:
    os.makedirs(os.path.join(ROOT, 'dataset/lang8/'))
except:
    pass
# 把lang8原始数据放到该目录下
lang8_data = open(os.path.join(ROOT, 'dataset/lang8/lang8.train.ccl22.para'), 'r').readlines()
print(len(lang8_data))
```

```python
# 数据清洗函数
import re
import emoji

from opencc import OpenCC
def Tradition2Simply(line):
    # 繁体转简体
    cc = OpenCC('t2s')
    line = cc.convert(line)
    return line

def rmText(line):
    # 清洗特殊字段
    rm_text_list1 = ['www', '｜', '^_^', 'T_T', '+_+', '-_-', '=_=', '@.@']
    for text in rm_text_list1:
        line = line.replace(text, '')
    return line

def rmBiaoDian(line):
    # 清洗重复标点
    pattern = re.compile(r'(,){2,}')
    line = pattern.sub(',', line.strip())

    pattern = re.compile(r'("){2,}')
    line = pattern.sub('"', line.strip())

    pattern = re.compile(r"('){2,}")
    line = pattern.sub("'", line.strip())

    pattern = re.compile(r"(——){1,}")
    line = pattern.sub("", line.strip())

    pattern = re.compile(r'(、){2,}')
    line = pattern.sub('、', line.strip())

    pattern = re.compile(r'(“){2,}')
    line = pattern.sub('“', line.strip())

    pattern = re.compile(r'(”){2,}')
    line = pattern.sub('”', line.strip())

    pattern = re.compile(r'(w){3,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(\\){2,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(@){1,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(\#){1,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(\$){2,}')
    line = pattern.sub('$', line.strip())
    # print(line)

    pattern = re.compile(r'(\^){1,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(&){2,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(%){2,}')
    line = pattern.sub('%', line.strip())
    # print(line)

    pattern = re.compile(r'(\(){2,}')
    line = pattern.sub('(', line.strip())
    # print(line)

    pattern = re.compile(r'(\)){2,}')
    line = pattern.sub(')', line.strip())
    # print(line)

    pattern = re.compile(r'(-){1,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(，){2,}')
    line = pattern.sub('，', line.strip())
    # print(line)

    pattern = re.compile(r'(。){2,}')
    line = pattern.sub('。', line.strip())
    # print(line)

    pattern = re.compile(r'(;){2,}')
    line = pattern.sub(';', line.strip())
    # print(line)

    pattern = re.compile(r'(\?){2,}')
    line = pattern.sub('?', line.strip())
    # print(line)

    pattern = re.compile(r'(？){2,}')
    line = pattern.sub('？', line.strip())
    # print(line)

    pattern = re.compile(r'(！){2,}')
    line = pattern.sub('！', line.strip())
    # print(line)

    pattern = re.compile(r'(!){2,}')
    line = pattern.sub('!', line.strip())
    # print(line)

    pattern = re.compile(r'(:){2,}')
    line = pattern.sub(':', line.strip())
    # print(line)

    pattern = re.compile(r'(；){2,}')
    line = pattern.sub('；', line.strip())
    # print(line)

    pattern = re.compile(r'(\[){2,}')
    line = pattern.sub('[', line.strip())
    # print(line)

    pattern = re.compile(r'(\]){2,}')
    line = pattern.sub(']', line.strip())
    # print(line)

    pattern = re.compile(r'({){2,}')
    line = pattern.sub('{', line.strip())
    # print(line)

    pattern = re.compile(r'(}){2,}')
    line = pattern.sub('}', line.strip())
    # print(line)

    pattern = re.compile(r'(【){2,}')
    line = pattern.sub('【', line.strip())
    # print(line)

    pattern = re.compile(r'(】){2,}')
    line = pattern.sub('】', line.strip())
    # print(line)

    pattern = re.compile(r'(\.){2,}')
    line = pattern.sub('.', line.strip())
    # print(line)

    pattern = re.compile(r'(\*){1,}')
    line = pattern.sub('', line.strip())
    # print(line)

    # pattern = re.compile(r'(/){2,}')
    # line = pattern.sub('/', line.strip())
    # print(line)

    pattern = re.compile(r'(<){2,}')
    line = pattern.sub('<', line.strip())
    # print(line)

    pattern = re.compile(r'(>){2,}')
    line = pattern.sub('>', line.strip())
    # print(line)

    pattern = re.compile(r'(_){1,}')
    line = pattern.sub('', line.strip())

    pattern = re.compile(r'(=){1,}')
    line = pattern.sub('', line.strip())

    pattern = re.compile(r'(\+){1,}')
    line = pattern.sub('', line.strip())
    # print(line)

    pattern = re.compile(r'(\|){1,}')
    line = pattern.sub('', line.strip())

    pattern = re.compile(r'(\~){1,}')
    line = pattern.sub('', line.strip())
    # print(line)
    return line

def rmEmoji(line):
    # 清洗emoji表情
    line = emoji.replace_emoji(line.strip(), replace="")
    return line
```

```python
# 清洗数据
from multiprocessing import cpu_count, Pool
lang8_data_ = []
def multi_process(line):
    line = Tradition2Simply(rmBiaoDian(rmEmoji(rmText(line.strip().lower())))) + '\n'
    return line
pool = Pool(cpu_count())
# 数据分发 合并
lang8_data_ = pool.map(multi_process, lang8_data) # 把数据，和要处理数据的函数，分别导入对应的进程池
# 关闭进程池
pool.close()
# 执行完close后不会有新的进程加入到pool,join函数等待所有子进程结束
pool.join()

open(os.path.join(ROOT, 'dataset/lang8/lang8_clean_data.txt'), 'w').write(''.join(lang8_data_))
```

```python
def CountChinese(line):
    # 统计中文字符
    cnt = 0
    for ch in line:
        if '\u4e00' <= ch <= '\u9fff':
            # return True
            cnt += 1
    return cnt
```

```python
# 针对清洗后的lang8数据，去除异常样本
import tqdm
tmp1, tmp2, tmp3 = [], [], []
src_data, trg_data = [], []
lang8_data = open(os.path.join(ROOT, 'dataset/lang8/lang8_clean_data.txt')).readlines()
for line in tqdm.tqdm(lang8_data, total=len(lang8_data)):
    line = line.strip().split('\t')
    assert len(line) == 2
    src, trg = line[0].replace(' ', ''), line[1].replace(' ', '')
    if abs(len(src)-len(trg)) >= 6:
        tmp1.append(src+'\t'+trg+'\n')
        continue
    if len(src) <= 5 or len(trg) <= 5:
        tmp2.append(src+'\t'+trg+'\n')
        continue
    if CountChinese(src)/len(src)<0.2 or CountChinese(trg)/len(trg)<0.2:
        # print(src, trg)
        tmp3.append(src+'\t'+trg+'\n')
        continue

    src_data.append(src)
    trg_data.append(trg)
assert len(src_data) == len(trg_data)
open(os.path.join(ROOT, 'dataset/lang8/lang8异常样本(长度差别>=6).txt'), 'w').write(''.join(tmp1))
open(os.path.join(ROOT, 'dataset/lang8/lang8异常样本(文本过短<=5).txt'), 'w').write(''.join(tmp2))
open(os.path.join(ROOT, 'dataset/lang8/lang8异常样本(中文字符占比<20%).txt'), 'w').write(''.join(tmp3))
open(os.path.join(ROOT, 'dataset/lang8/lang8_src.txt'), 'w').write('\n'.join(src_data))
open(os.path.join(ROOT, 'dataset/lang8/lang8_trg.txt'), 'w').write('\n'.join(trg_data))
len(src_data)
```

```python
# 将编辑文件处理为训练集jsonl
import json

input_file, truth_file = os.path.join(ROOT, 'dataset/lang8/lang8_src.txt'), os.path.join(ROOT, 'dataset/lang8/lang8_edits.txt')
correct_file = os.path.join(ROOT, 'dataset/lang8/lang8_trg.txt')


inputs = open(input_file, 'r').readlines()
truths = open(truth_file, 'r').readlines()
correct = open(correct_file, 'r').readlines()
print(inputs[0].strip())
print(truths[0].strip().split(',\t'))
print(correct[0].strip())
item = {'text': '', 'correct': ''}
res = {}
for index, line in enumerate(inputs):
    line = line.strip()
    item['text'] = line
    item['correct'] = correct[index].strip()
    if str(index) not in res:
        res[str(index)] = item
    else:
        print('重复的输入文本', line)
    item = {'text': '', 'correct': ''}
print(len(res))
for line in truths:
    line = line.strip().split(',\t')
    try:
        assert len(line) == 2 or len(line) == 4 or len(line) == 5
    except:
        print('长度不匹配:', line)

    if 'error' not in res[line[0]]:
        res[line[0]]['error'] = []

    if len(line) == 5 and line[-2] == 'M' and line[-1] in ['。', '?', '，', '.', '？', '!', '！', ';', '；', '"', '“', '”', "'", '‘', '’', '、'] and int(line[-3]) >= len(res[line[0]]['text'])+1:
        # 去除由于末尾缺失标点带来的“M”类型错误，并补上缺失的标点
        print(res[line[0]]['text'], '/', res[line[0]]['correct'], line)
        res[line[0]]['text'] += line[-1]
        print(res[line[0]]['text'])
        print()
        continue

    if len(line) != 2:
        res[line[0]]['error'].append({'start': int(line[1])-1, 'end': int(line[2]), 'type': line[3], 'correct': line[4] if len(line)==5 else ''})
for k, v in res.items():
    if 'error' not in v:
        print('没有错误: ', k, v)
# res
result = []
for k, v in res.items():
    if 'error' in v:
        result.append({'file_name': input_file, 'id': k, 'text': v['text'], 'correct': v['correct'], 'error': v['error'][:]})
print(len(result))
open(os.path.join(ROOT, 'dataset/lang8/lang8_data_v1.jsonl'), 'w').write('\n'.join(json.dumps(item, ensure_ascii=False) for item in result))
```
